**Proyecto Calculo integral, Aproximacion integrales metodo trapecio**

El programa aplica la formula del trapecio a una funcion proporcionada por el usuario dividiendo entre el intervalo   a,b proporcionados por el usuario

```python
import numpy as np
import matplotlib.pyplot as plt


def trapecio(f, a, b, n):
    h = (b - a) / n
    suma = f(a) + f(b)
    for i in range(1, n):
        suma += 2 * f(a + i * h)
    return (h / 2) * suma


print("Método del Trapecio")
expr = input("Función f(x): ").strip()
f = lambda x: eval(expr)
a = float(input("Límite inferior a: "))
b = float(input("Límite superior b: "))
n = int(input("Número de trapecios n: "))

resultado = trapecio(f, a, b, n)
print(f"Resultado: {resultado:.8f}")

# Gráfica
x = np.linspace(a, b, 500)
y = np.vectorize(f)(x)
plt.plot(x, y, color="blue", linewidth=2)

h = (b - a) / n
for i in range(n):
    x0, x1 = a + i * h, a + (i + 1) * h
    y0, y1 = f(x0), f(x1)
    plt.fill_between([x0, x1], [y0, y1], alpha=0.3, color="orange")
    plt.plot([x0, x0], [0, y0], color="orange", linewidth=0.8)
    plt.plot([x1, x1], [0, y1], color="orange", linewidth=0.8)
    plt.plot([x0, x1], [y0, y1], color="orange", linewidth=0.8)

plt.title("Método del Trapecio")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

"""
x²          -> x**2
√x          -> np.sqrt(x)
e^x         -> np.exp(x)
ln(x)       -> np.log(x)
sen(x)      -> np.sin(x)
cos(x)      -> np.cos(x)
"""
```
